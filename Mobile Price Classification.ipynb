```python
from pycaret.classification import *
```

```python
clf = setup(data=df_train, target = "price_range", session_id=123, normalize= True)
```

```python
evaluate_model(best)
```

```python
plot_model(best, plot = 'confusion_matrix')
```

```python
plot_model(best, plot = 'auc')
```

```python
predict_model(best)
```
